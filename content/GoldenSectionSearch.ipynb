```python
# En ne réutilisant pas un des points milieu:

def goldenSearchSimple( f, a, b ):
    epsilon = .001
    rho = 0.3
    
    while b-a > epsilon:
        m1 = rho*(b-a) + a
        m2 = b - rho*(b-a)
        f1, f2 = f( m1 ), f( m2 )  
    
        if f1 < f2:
            a, b = a, m2
        else:
            a, b = m1, b
            
        #print a, b

    xZero = (a+b)/2.0 
    return xZero
```

```python
# En réutilisant un des points milieu:

def goldenSearch( f, a, b ):
    epsilon = .001
    rho = 0.382
    
    m1 = rho*(b-a) + a
    m2 = b - rho*(b-a)
    f1, f2 = f( m1 ), f( m2 )  
    
    while b-a > epsilon:
        
        if f1 < f2:
            a, b = a, m2
            m2, f2 = m1, f1
            m1 = rho*(b-a) + a
            f1 = f( m1 )
        else:
            a, b = m1, b
            m1, f1 = m2, f2
            m2 = b - rho*(b-a)
            f2 = f( m2 )
            
        #print a, b

    xZero = (a+b)/2.0 
    return xZero
```

### Tests

```python
f = lambda X: .2*np.sin(24*X) + 5*(2*X-0.8)**2
xMin = goldenSearchSimple( f, 0, 1 )   

X = np.linspace( 0, 1, 51 )
Y = f( X )

plt.figure(figsize=(12, 4)  )
plt.plot( X, Y,  'k-', alpha=0.7 )
plt.plot( xMin, f( xMin ),  'ro', alpha=0.7 )
```
