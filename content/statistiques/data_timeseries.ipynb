```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
import os
datafiles = os.listdir('./data')

datafiles = [ './data/' + name for name in datafiles ]
print datafiles
```

```python
def readXY( filename ):
    data = np.loadtxt(filename, dtype='str', comments='#', delimiter=',',  skiprows=1)

    Y = np.array( [ float(u[1]) for u in data ])
    X = np.array( range(len(Y)) )
    
    return X, Y
```

```python
for fname in datafiles:
    X, Y = readXY( fname )
    
    print fname
    plt.figure(figsize=(12, 3)  )
    plt.plot( X, Y,  'k-', alpha=0.5 );
    plt.title( fname )
```

## Save to a file

```python
# Save

data = {}
for fname in datafiles:
    X, Y = readXY( fname )
    
    print fname, len(Y)
    data[fname] = (X, Y)
    
np.save('./timeseriesdata.npy', data) 
```
