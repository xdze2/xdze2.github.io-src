```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
import os
datadir = './data'
datafiles = os.listdir(datadir)

datafiles = [ name for name in datafiles ]
print( datafiles )
```

```python
def readXY( filename ):
    data = np.loadtxt(filename, dtype='bytes', comments='#', delimiter=',',  skiprows=1)

    Ystr = [ u[1] for u in data ]
    Y = []
    for i, s in enumerate( Ystr ):
        if s != b'.':  # on supprime les données manquantes
            Y.append( float(s) )
            
    Y = np.array( Y )
    X = np.array( range(len(Y)) )
    
    return X, Y
```

```python
for fname in datafiles:
    X, Y = readXY( datadir + '/' + fname )
    
    print( fname )
    plt.figure(figsize=(12, 3)  )
    plt.plot( X, Y,  'k-', alpha=0.5 );
    plt.title( fname )
```

## Save to a file

```python
# Save

data = {}
for fname in datafiles:
    X, Y = readXY( datadir + '/' + fname )
      
    print( '%s - %i'% (fname, len(Y)) )
    data[fname] = (X, Y)
    
np.save('./timeseriesdata.npy', data) 
```
