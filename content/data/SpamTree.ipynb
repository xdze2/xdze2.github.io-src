```python
df = pd.read_csv('./files/spam.data', sep=' ', header=0, names=names['col'])
df['train'] = trainset
```

```python
Tr = df[df['train']==1]
Tr = Tr.drop(['train'], axis=1)
```

```python
variable = 'george'
seuil = .4
```

```python
def Gini( seuil, variable ):
    p = ( Tr[Tr[variable]<seuil]['is_spam']==1 ).sum()/( Tr[variable]<seuil ).sum()
    gini = 2*p*(1-p)
    return gini

#Â on cherche seuil donnant un Gini maximal
```

```python
Gini(1000, variable)
```

```python
from scipy.optimize import minimize
from scipy.optimize import fminbound
```

```python
res = minimize(lambda s: -Gini(s[0], variable), Tr[variable].max())
res
```

```python
res = fminbound(lambda s: -Gini(s, variable), 0, Tr[variable].max() )
print(res)
print( Gini(res, variable) )
```

```python
def gini2( F, S, seuil ):
    nspam = S[ F<seuil ].sum()
    ntot = len( S[ F<seuil ] )
    g = nspam*(1-nspam/ntot)

    nspam = S[ F>=seuil ].sum()
    ntot = len( S[ F>=seuil ] )
    g += nspam*(1-nspam/ntot)
    
    return g
```
